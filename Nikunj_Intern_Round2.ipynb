# Task1: Visualization

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Data Of Performance ScoreCard

```python
df = pd.read_excel('Performance_intern.xlsx')
df
```

# Line Chart to compare weekly Performance

```python
# Get unique performance categories
performances = df['Performance'].unique()

# Create a separate line plot for each performance category
for performance in performances:
    subset = df[df['Performance'] == performance]
    x = range(1, len(subset) + 1)  # Set x-axis values as 1, 2, 3
    plt.plot(x, subset['Neha'], marker='o', label='Neha')
    plt.plot(x, subset['Amit'], marker='*', label='Amit')
    plt.plot(x, subset['Sachin'], marker='+', label='Sachin')

    # Set labels and title for each plot
    plt.xlabel('Week')
    plt.title(f'Performance Scores for {performance}')

    # Add legend
    plt.legend()

    # Show the plot
    plt.show()
```

# Pie Chart to compare Total Performance

```python
# Create a pie chart for each performance and candidate
for performance in performances:
    plt.figure()
    
    # Filter data for the specific performance category
    performance_data = df[df['Performance'] == performance]
    
    # Calculate the total count for each candidate
    candidate_counts = performance_data.drop('Performance', axis=1).sum()
    
    # Create the pie chart
    plt.pie(candidate_counts, labels=candidate_counts.index, autopct='%1.1f%%', startangle=90)
    
    # Add a title
    plt.title(f'Distribution of {performance}')
    
    # Display the chart
    plt.show()
```
